```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from scipy.cluster.hierarchy import dendrogram,linkage

iris = load_iris()
data = iris.data[:6]
def proximity(data):
    n = data.shape[0]
    matrix = np.zeros((n,n))
    
    for i in range(n):
        for j in range(i+1,n):
            matrix[i][j] = np.linalg.norm(data[i]-data[j])
            matrix[j][i] = matrix[i][j]
            
    return matrix

def plotdendro(data,method):
    linkagemat = linkage(data,method=method)
    dendrogram(linkagemat)
    plt.title(f"{method} linkage")
    plt.show()
    
print(proximity(data))
plotdendro(data,'single')
plotdendro(data,'complete')
            
```
